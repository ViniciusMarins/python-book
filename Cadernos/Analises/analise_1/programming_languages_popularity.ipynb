# Most Popular Programming Languages Since 2004

- Popularity of Programming Languages from 2004 to 2023
- Dataset Font (Kaggle): https://www.kaggle.com/datasets/muhammadkhalid/most-popular-programming-languages-since-2004/
- Este conjunto de dados contém informações sobre as Linguagens de Programação Mais Populares de 2004 a 2023. Todos os valores das Linguagens de Programação estão em forma de porcentagem em relação a 100 %.


## Nota: Dataset

- O Dataset original conta com 30 colunas, sendo elas 29 para linguagens de programação e uma referente a data, foi realizado uma manipulação na tabela para que apenas as linguagens tratadas abaixo fossem utilizadas. 
- Com isso, o somatório total dos valores percentuais de todas as linguagens apresentados são referente à uma amostra da tabela original, ou seja, eles resultam em menos de 100%.


Foi realizado a inserção de uma nova coluna "Year" baseada na coluna "Date" já existente no dataset. A nova inserção foi necessária para converter o formato de perído de [mês ano] para [ano], criando apenas uma coluna com os anos.
Após o dataset ser manipulado corretamente, foi realizado um agrupamento dos dados por ano.

```python
import pandas as pd
import matplotlib.pyplot as plt

path_dataset = "dataset.csv"
dados = pd.read_csv(path_dataset)

dados['Year'] = dados['Date'].str.split(' ').str[-1].astype(int)

colunas = ['Year','Delphi/Pascal','Matlab','C/C++','C#','PHP','Python','R','TypeScript','Java','JavaScript']
new_dados = dados[colunas]

dados = new_dados.groupby('Year').mean()

dados = dados.applymap(lambda x:round(x,2))

display(dados)
```

```python
# Função para criação dos gráficos

def grafico_per_year(ano_desejado):
    dados_ano = dados.loc[ano_desejado]

    dados_ano = dados_ano.reset_index().rename(columns={ano_desejado: 'Popularidade'})
    dados_ano = dados_ano.rename(columns={'index': 'Linguagem'})

    plt.figure(figsize=(10, 6))

    dados_ano.plot(kind='bar', x='Linguagem', y='Popularidade', stacked=True, legend=True)
    plt.xlabel('Linguagem')
    plt.ylabel('Popularidade')
    plt.title(f'Popularidade das Linguagens de Programação em {ano_desejado}')
    plt.xticks(rotation=45)

    plt.show()

def grafico_overal():
    dados_transpostos = dados.transpose()

    plt.figure(figsize=(12, 6))

    for linguagem in dados_transpostos.index:
        plt.plot(dados_transpostos.columns, dados_transpostos.loc[linguagem], label=linguagem)

    plt.xlabel('Ano')
    plt.ylabel('Popularidade')
    plt.title('Popularidade das Linguagens de Programação ao Longo dos Anos')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.xticks(dados_transpostos.columns.astype(int), rotation=45)

    plt.show()
```

```python
# Plot dos Gráficos
grafico_overal()
grafico_per_year(2004)
grafico_per_year(2023)
```
